Úkol 1
V souboru s daty máš data o hodnotách finančních indikátorů 100 největších společností obchodovaných na americké burze. Naším cílem je zjistit, které indikátory nejvíce ovlivňují cenu, a vytvořit model, který odhadne cenu akcie na základě hodnot finančních indikátorů.

```python
import pandas as pd
import seaborn as sns
from scipy import stats
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import OLSInfluence
import numpy as np

data = pd.read_csv("df.csv")
data.head()
```

```python
data_corr = data.drop(columns=["Company", "Sector"])
data_corr.head()
```

Část 1
Vytvoř korelační matici a podívej se, který ukazatel má největší vliv na indikátor Y akcie.

Část 2
Vytvoř regresní model, který bude mít koeficient Y jako vysvětlovanou proměnnou. Do modelu vlož hodnoty ostatních indikátorů jako vysvětlující proměnné.
S využitím modulu statsmodels vytvoř regresní model a zobraz si tabulku se souhrnem významů. 

```python
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"
mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```

Podívej se na hodnoty koeficientů a na výsledky testu statistické významnosti koeficientů. Pokud je některý koeficient (nebo více koeficientů) nevýznamný, sestav nový model bez tohoto koeficientu (případně beze všech nevýznamných koeficientů).

--> CR, DA, RT, DE a PT jsou statisticky nevýznamné (p hodnota je větší než 0.05)

```python
formula = "Y ~ FL + OPM + TAT"
mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```

Pro všechna data odhadni ukazatel Y s využitím tvého modelu a odhadnuté ceny vlož do původní tabulky s daty. Dále vypočítej rozdíl mezi odhadem koeficientu a jeho skutečnou hodnotou.

```python
data["Estimated Y"] = res.predict(data)
data["Variance"] = data["Y"] - data["Estimated Y"]
data.head()
```

Najdi akcii, kde je tento rozdíl největší (tj. hledáme akcii, které náš model predikuje výrazně vyšší cenu, než jaká je ve skutečnosti, tato akcie je potenciálně na trhu podhodnocená).

```python
min_var = data["Variance"].min()
data[data["Variance"] == min_var]
```

Bonus 1: Sestav model s využitím robustní regrese. Opět proveď vyřazení koeficinetů, které nejsou statisticky významné, a sestav model pouze s významnými koeficienty. Vlož odhady cen do původních dat a opět najdi potenciálně nejvíce podhodnocenou akcii. Jde v případě robustní regese o stejnou akcii, nebo se akcie liší?

```python
data_robust = pd.read_csv("df.csv")
```

```python
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"
mod = smf.rlm(formula=formula, data=data_robust)
res = mod.fit()
res.summary()
```

```python
formula = "Y ~ CR + FL + OPM + TAT"
mod = smf.rlm(formula=formula, data=data_robust)
res = mod.fit()
res.summary()
```

```python
data_robust["Estimated Y"] = res.predict(data_robust)
data_robust["Variance"] = data_robust["Y"] - data_robust["Estimated Y"]
min_var = data_robust["Variance"].min()
data_robust[data_robust["Variance"] == min_var]
```

Je to stejná společnost.

Bonus 2: Použij původní (tedy "nerobustní") model a vyčísli Cookovu vzdálenost pro všechny hodnoty. Vyřaď všechny akcie s Cookovou vzdáleností vyšší než 1 a přepočítej regresní model. Nakonec opět najdi potenciálně nejvíce podhodnocenou akcii a podívej se, jestli jde o stejnou akcii jako u předchozích modelů.

```python
data_cook = pd.read_csv("df.csv")
```

```python
formula = "Y ~ FL + OPM + TAT"
mod = smf.ols(formula=formula, data=data_cook)
res = mod.fit()

influence = OLSInfluence(res)
data_cook["Cook Distance"] = influence.cooks_distance[0]
#data["Cook Distance"].max()
```

```python
data_cook = data_cook[data_cook["Cook Distance"] < 1]
```

```python
mod = smf.ols(formula=formula, data=data_cook)
res = mod.fit()

data_cook["Estimated Y"] = res.predict(data_cook)
data_cook["Variance"] = data_cook["Y"] - data_cook["Estimated Y"]
min_var = data_cook["Variance"].min()
```

```python
data_cook[data_cook["Variance"] == min_var]
```

Jde o stejnou společnost. V datech nebyla Cookova vzdálenost nikde větší než jedna.

Bonus 3: Přidej Sektor (poslední sloupec) jako vysvětlující proměnnou s využitím One Hot Encoding. Podívej se, které sektory mají kladný koeficient a které sektory mají záporný koeficient.

```python
data_sector = pd.read_csv("df.csv")
```

```python
data_sector["Sector"].value_counts()
```

```python
data_sector = pd.get_dummies(data=data_sector, prefix = "is", columns=["Sector"], dtype=int)
data_sector.head()
```

```python
data_sector = data_sector.rename(columns={"is_Basic Materials":"is_Bas_Materials","is_Communication Services": "is_Com_Services", "is_Consumer Cyclical": "is_Con_Cyclical", "is_Consumer Defensive": "is_Con_Defensive","is_Financial Services": "is_Fin_Services"})
data_sector.columns
```

```python
formula = "Y ~ FL + OPM + TAT + is_Bas_Materials  + is_Com_Services + is_Conglomerates + is_Con_Cyclical + is_Con_Defensive + is_Energy + is_Fin_Services + is_Healthcare + is_Industrials + is_Technology"
mod = smf.ols(formula=formula, data=data_sector)
res = mod.fit()
res.summary()
```

Sektory s kladným koeficientem: Basic Materials, Consumer Cyclical, Financial Services, Technology

Sektory se záporným koeficientem: Communication Services, Conglomerates, Consumer Defensive, Energy, Healthcare, Industrials

Statisticky významný se zdá být pouze Industrials. Některé kategorie jsou ale velmi málo zastoupené.
